```python
import os
import pandas as pd
import uuid
import csv
from datetime import datetime


# 1. read csv and make the first row the column
df = pd.read_csv('Hadsundvej.csv', error_bad_lines=False, sep=',', skiprows=1)

#set column names
df = df.set_axis(['PlaceID', 'Direction','RouteID', 'Traffic Count', 'Days', 'Traffic Type', 'Date'], axis=1)
# df = df.drop([0])

#convert date into string and create datetime column
df['Date'] = df['Date'].astype(str)
df[['Date1', 'Date2']] = df['Date'].str.split(pat="-", expand = True)
df.head()
df['Datetime'] = pd.to_datetime(df['Date1'], format='%d.%m.%y')
df = df.drop(['Date1', 'Date2'], axis=1)

 #create new column with months and years
# https://stackoverflow.com/questions/24082784/pandas-dataframe-groupby-datetime-month
df['YearMonth'] = pd.to_datetime(df['Datetime']).apply(lambda x: '{year}-{month}'.format(year=x.year, month=x.month))

#write each YearMonth into new csv
for i in df ['YearMonth']:
    df_yearmonth = df[df['YearMonth'] == i]
    df_yearmonth.to_csv(f"Hadsundvej{i}.csv")
```
